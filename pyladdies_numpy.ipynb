# TRABAJO CON IMAGENES USANDO NUMPY Y SKIMAGE
Se proponen ejemplos simples para empezar con el trabajo con matrices de imágenes

```python
# Librerías para el trabajo de imágenes
from skimage import io, img_as_float, data
from skimage.color import rgb2gray
```

```python
image1 = io.imread("https://s3-us-west-2.amazonaws.com/publicfilesm/maestriaIa/Imagen1.png");
print( io.imshow(image1,) );
print( image1.shape ); #Se tiene una imagen con 4 canales de colores
```

```python
#image1
```

```python
image1 = img_as_float(image1);
# Construyendo a color de imagen usando solamente un canal
r = image1[:, :, 0];
g = image1[:, :, 1];
b = image1[:, :, 2];

red_image = np.zeros((197, 200, 3));
red_image[:, :, 0] = r;
red_image[:, :, 1] = np.zeros((197, 200));
red_image[:, :, 2] = np.zeros((197, 200));

green_image = np.zeros((197, 200, 3));
green_image[:, :, 0] = np.zeros((197, 200));
green_image[:, :, 1] = g;
green_image[:, :, 2] = np.zeros((197, 200));

blue_image = np.zeros((197, 200, 3));
blue_image[:, :, 0] = np.zeros((197, 200));
blue_image[:, :, 1] = np.zeros((197, 200));
blue_image[:, :, 2] = b;

fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(16, 10), sharex=True, sharey=True);
ax1.imshow(red_image);
ax1.grid(False)
ax2.imshow(green_image);
ax2.grid(False)
ax3.imshow(blue_image);
ax3.grid(False)
```

```python
image1_gris = rgb2gray(image1); #Se pasa a gris la imagen

# Se va a utilizar técnicas sencillas para rotar la matriz imagen
# primero cambiamos el orden de las filas en la matriz, ubicando la última a primera, etc
image_rot = image1_gris[::-1].copy();
#Luego creamos una variable temporal de trabajo
temp_row = image_rot.copy();

#Finalmente cambiamos el orden de las columnas enviado la primera a última, etc
for i in range(image_rot.shape[1]):
    image_rot[:, i] = temp_row[:,image_rot.shape[1]-1-i];
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,figsize=(10, 5))

ax1.imshow(image1_gris)
ax1.axis('off')

ax2.imshow(image_rot)
ax2.axis('off')

plt.show()
print( image1_gris.shape )
```

```python
#Mapas de una imagen
plt.imshow(image1_gris[:,:],cmap = 'hsv')
```

```python
coins = data.coins()
type(coins)
```

```python
plt.imshow(coins,cmap='gray')
plt.grid(False)
plt.show()
```

```python
max_per_row = np.max(coins, axis=1);
print(max_per_row);
plt.plot(max_per_row);
plt.grid(False)
plt.show()
```

```python
umbral = 160;
binary_max_per_row = 255 * (max_per_row > umbral) #Si ubico max_per_row > umbral, devuelve 1 si es cierto 0 si falso

plt.plot(binary_max_per_row);
plt.grid(False)
plt.show()
```

```python
# Copia de la imagen para realizar pruebas
test_coins = coins.copy();

# Todos aquellas posiciones del vector que tengan valores que iguales a 0 los modifico en la imagen poniéndolos
# también a 255.
test_coins[binary_max_per_row == 0, :] = 255;

plt.imshow(test_coins,cmap='gray')
plt.grid(False)
plt.show()
```

De esta manera, me quedo con un acotamiento de dónde están las monedas. Este tipo de estrategias sirven para detectar objetos alineados (como puede ser la información dentro de un DNI).
Si repetimos la misma operación para las componentes verticales, veamos qué sucede.

```python
#Máximo por columnas
max_per_columns = np.max(coins, axis=0);

#Se representa la gráfica de los máximos
plt.plot(max_per_columns);
plt.show()
```

```python
umbral2 = 200;
binary_max_per_columns = 255 * (max_per_columns > umbral2);
plt.plot(binary_max_per_columns);
plt.show()
```

Aquí el umbral es más complicado de elegir. En este caso, podemos ver que existen 6 mínimos, correspondientes a las 6 monedas que tenemos por cada fila.
Se puede decir que estos mínimos están, aproximadamente (que es lo que buscamos ahora), en las posiciones: 75, 130, 180, 240, 290.
Por supuesto, estos mínimos se pueden obtener de forma más elegante con un algoritmo de detección de mínimos locales (from scipy.signal import argrelextrema)

```python
local_minima = np.array([75,130,180,240,290]);
local_minima
```

```python
test_coins[:, local_minima] = 255

plt.imshow(test_coins,cmap='gray')
plt.grid(False)
plt.show()
```
